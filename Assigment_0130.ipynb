```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://github.com/raraden/SG3/raw/main/accident.csv"
df = pd.read_csv(url)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(df)
```

```python
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['pink', 'blue'])
plt.title("Grafik Kecelakaan Berdasarkan Jenis Kelamin")
plt.show()
```

Insight dari data diatas adalah kebanyakan korban kecelakaan adalah wanita dengan presentase sebesar 55%.

```python
seatbelt_counts = df["Seatbelt_Used"].value_counts()
helmet_counts = df["Helmet_Used"].value_counts()

plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.pie(seatbelt_counts, labels=seatbelt_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'orange'])
plt.title("Penggunaan Seatbelt pada Korban Kecelakaan")
plt.subplot(1, 2, 2)
plt.pie(helmet_counts, labels=helmet_counts.index, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title("Korban Kecelakaan yang Menggunakan Helm")
plt.tight_layout()
plt.show()
```

Dari grafik diatas, kita mendapatkan informasi bahwa sebanyak 55 persen korban kecelakaan menggunakan helm, dan sebanyak 55 persen korban kecelakaan menggunakan seatbelt. Datasheet diatas tidak menampilkan korban kecelakaan menggunakan kendaraan apa.

```python
df_numeric = df.select_dtypes(include=['number'])
mean_values = df_numeric.mean()
mode_values = df_numeric.mode().iloc[0]

print("Rata-rata dari tiap kolom numerik:\n")
print(mean_values.to_string(index=True))

print("\nModus dari tiap data:\n")
print(mode_values.to_string(index=True))
```

Insight yang didapat adalah rata-rata orang mengalami kecelakaan  berumur disektiar 43 tahun dengan kecepatan rata-rata di 70 km/jam dan sekitar 50 persen orang dapat bertahan hidup. Kebanyakan orang mengalami kecelakaan di umur 56 tahun dengan kecepatan terbanyak di 77 km/jam.
